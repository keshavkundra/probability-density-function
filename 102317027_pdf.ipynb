```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
data = pd.read_csv("data.csv", encoding="latin1")

print(data.columns)
```

```python
x = data["no2"]

# Remove missing values
x = x.dropna().values

print("Total samples:", len(x))
```

```python
r = 102317027

ar = 0.05 * (r % 7)
br = 0.3 * ((r % 5) + 1)

print(ar, br)

```

```python
z = x + ar * np.sin(br * x)
```

```python
mu = np.mean(z)
sigma = np.std(z)

print("mu =", mu)
print("sigma =", sigma)
```

```python
lam = 1 / (2 * sigma**2)
lam
```

```python
c = 1 / np.sqrt(2 * np.pi * sigma**2)
c
```

```python

mu = np.mean(z)
sigma = np.std(z)

lam = 1 / (2 * sigma**2)
c = 1 / np.sqrt(2 * np.pi * sigma**2)

print("\nFINAL ANSWERS")
print("mu =", mu)
print("lambda =", lam)
print("c =", c)
```
